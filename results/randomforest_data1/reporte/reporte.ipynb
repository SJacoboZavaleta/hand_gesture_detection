```python
import pickle
import pandas as pd

# Cargar el archivo pickle
with open('hand_gesture_detection/src/randomForest/unified_webcam_dataset_info.pickle', 'rb') as f:
    data = pickle.load(f)

# Imprimir el tamaño del conjunto de datos
print(f"Dataset size: {len(data)}")

# Comprobar que el dataset no está vacío
if len(data) > 0:
    print("First 5 rows:")
    # Convertir los datos a un DataFrame de Pandas
    df = pd.DataFrame(data)
    # Mostrar las primeras 5 filas del DataFrame
    print(df.head(5))
else:
    print("The dataset is empty.")

print(df['labels'].value_counts().sum())
print(len(df['data'][0]))
print(df['labels'].head(5))
print(df['data'][df['labels']=='28'].tail(5))
```
